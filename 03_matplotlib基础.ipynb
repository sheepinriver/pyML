```python
np.random.seed(666)
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(8,4)) # - 

#
plt.scatter(x, y
            , s=5 #
            , c="b" #
            , alpha=0.5
            , label="Scatter Figure"
           )

#plt.xticks(np.arange(-3,3.5,0.5))
plt.tick_params(direction="out"
                , length=6
                , width=2
                , colors="w"
               # , grid_color='r'
               # , grid_alpha=0.5
               )

#plt.legend() #
plt.show()
```

```python
plt.plot(x, y, color="red", linestyle="--")
plt.axis([-1, 11, -2, 2])
#plt.xlim(-1, -1)
plt.xlabel("x axis")
plt.ylabel("y value")
plt.tick_params(direction="out"
                , length=6
                , width=2
                , colors="w"
               # , grid_color='r'
               # , grid_alpha=0.5
               )
plt.show()
```
