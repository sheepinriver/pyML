```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
np.random.seed(666)
x1 = np.random.randn(1000)
x2 = np.random.randn(1000)
plt.figure(figsize=(8,4)) #实例一个画布 - 当只有一个图的时候，不是必须的

#绘图
plt.scatter(x1, x2
            , s=50 #散点大小
            , c="b" #散点颜色
            , alpha=0.5
            , label="Scatter Figure"
           )

#plt.xticks(np.arange(-3,3.5,0.5))
plt.tick_params(direction="out"
                , length=6
                , width=2
                , colors="w"
               # , grid_color='r'
               # , grid_alpha=0.5
               )

#plt.legend() #显示图例
plt.show()
```

```python
np.random.seed(666)
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(8,4)) #实例一个画布 - 当只有一个图的时候，不是必须的

#绘图
plt.scatter(x, y
            , s=5 #散点大小
            , c="b" #散点颜色
            , alpha=0.5
            , label="Scatter Figure"
           )

#plt.xticks(np.arange(-3,3.5,0.5))
plt.tick_params(direction="out"
                , length=6
                , width=2
                , colors="w"
               # , grid_color='r'
               # , grid_alpha=0.5
               )

#plt.legend() #显示图例
plt.show()
```

```python
plt.plot(x, y, color="red", linestyle="--")
plt.axis([-1, 11, -2, 2])
#plt.xlim(-1, -1)
plt.xlabel("x axis")
plt.ylabel("y value")
plt.tick_params(direction="out"
                , length=6
                , width=2
                , colors="w"
               # , grid_color='r'
               # , grid_alpha=0.5
               )
plt.show()
```

```python
np.random.seed(666)
x = np.linspace(0, 10, 100)
#y = np.cos(x)
plt.figure(figsize=(8,4)) #实例一个画布 - 当只有一个图的时候，不是必须的

#绘图
plt.scatter(x, np.sin(x)
            , s=5 #散点大小
            , c="b" #散点颜色
            , alpha=0.8
            , label="sin"
           )
plt.scatter(x, np.cos(x)
            , s=5 #散点大小
            , c="r" #散点颜色
            , alpha=0.5
            , label="cosin"
           )
#plt.xticks(np.arange(-3,3.5,0.5))
plt.tick_params(direction="out"
                , length=6
                , width=2
                , colors="w"
               # , grid_color='r'
               # , grid_alpha=0.5
               )

plt.legend() #显示图例
#plt.title("Welcome Henson !")
plt.show()
```
